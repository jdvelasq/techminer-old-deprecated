# Analysis: Co-ocurrence

```python
import pandas as pd
from techminer import RecordsDataFrame

rdf = RecordsDataFrame(
    pd.read_json(
        'step-07.json', 
        orient='records', 
        lines=True))
```

## Authors vs Source Title

```python
rdf.co_ocurrence(
    column_r='Authors', 
    column_c='Source title', 
    sep_r=',', 
    sep_c=None,
    top_n=20).tomatrix()
```

```python
rdf.co_ocurrence(
    column_r='Authors', 
    column_c='Source title', 
    sep_r=',', 
    sep_c=None,
    top_n=20).heatmap(figsize=(10,7))
```
