# Matrix Analysis

In this part, a second level analysis of the bibliographical information is done.

```python
import pandas as pd
```

```python
records = pd.read_json('demo-keywords.json', orient='records', lines=True)
```

## Keywords per year

```python
from techMiner import term_by_term
```

```python
term_by_term(
    df=records, 
    termA='Keywords', 
    termB='Year', 
    sepA=';', 
    sepB=None).head()
```

```python
term_by_term( 
    df=records, 
    termA='Keywords', 
    termB='Year', 
    sepA=';', 
    sepB=None, 
    minmax=(2,3)).head()   ## minimum and maximum values in the matrix
```

```python
from techMiner import heatmap

heatmap(
    term_by_term( 
        df=records, 
        termA='Keywords', 
        termB='Year', 
        sepA=';', 
        sepB=None, 
        minmax=(2,8)),
    ascendingA=False, ascendingB=False)
```

```python
from techMiner import alt_heatmap

alt_heatmap(
    term_by_term( 
        df=records, 
        termA='Keywords', 
        termB='Year', 
        sepA=';', 
        sepB=None, 
        minmax=(2,8)))
```

```python
from techMiner import alt_circleplot

alt_circleplot(
    term_by_term( 
        df=records, 
        termA='Keywords', 
        termB='Year', 
        sepA=';', 
        sepB=None, 
        minmax=(2,8)))
```

```python
m = term_by_term( 
    df=records, 
    termA='Keywords', 
    termB='Year', 
    sepA=';', 
    sepB=None, 
    minmax=(2,8))

m
```

```python
from techMiner import matrix
m = matrix(m)
m
```

```python
from techMiner import kmeans
centers, clusters = kmeans(m, n_clusters = 3)
```

```python
centers
```

```python
heatmap(centers, transform=False)
```

```python
clusters.sort_values('cluster')
```

```python
m.loc[clusters.sort_values('cluster').index,:]
```

## Autocorrelation

```python
from techMiner import correlation

correlation(
    df=records,
    termA='Keywords',
    sepA=';'
)
```

```python
alt_circleplot(
    correlation(
    df=records,
    termA='Keywords',
    sepA=';'))
```

```python
heatmap(
    correlation(
    df=records,
    termA='Keywords',
    sepA=';'))
```

## Crosscorrelation

```python
correlation(
    df=records,
    termA='Keywords',
    sepA=';',
    termB='Authors',
    sepB=','
)
```

```python
heatmap(
    correlation(
        df=records,
        termA='Keywords',
        sepA=';',
        termB='Authors',
        sepB=','))
```

```python
m = correlation(
    df=records,
    termA='Keywords',
    sepA=';',
    termB='Authors',
    sepB=','
)
m = matrix(m)
m
```

```python
centers, clusters = kmeans(m, n_clusters = 3)
```

```python
centers
```

```python
heatmap(centers, transform=False)
```

```python
m['cluster'] = clusters
m = m.sort_values('cluster')
m
```
