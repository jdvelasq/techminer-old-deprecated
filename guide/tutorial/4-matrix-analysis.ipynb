# Matrix Analysis

In this part, a second level analysis of the bibliographical information is done.

```python
import pandas as pd
```

```python
records = pd.read_json('demo-keywords.json', orient='records', lines=True)
```

## Keywords per year

```python
from techMiner import termByTerm
```

```python
termByTerm(
    df=records, 
    termA='Keywords', 
    termB='Year', 
    sepA=';', 
    sepB=None, 
    ascendingA=True, 
    ascendingB=True)
```

```python
termByTerm( 
    df=records, 
    termA='Keywords', 
    termB='Year', 
    sepA=';', 
    sepB=None, 
    ascendingA=True, 
    ascendingB=True, 
    minmax=(2,3))   ## minimum and maximum values in the matrix
```

```python
from techMiner import heatmap

heatmap(termByTerm( 
    df=records, 
    termA='Keywords', 
    termB='Year', 
    sepA=';', 
    sepB=None, 
    ascendingA=True, 
    ascendingB=True, 
    minmax=(2,3)))
```

```python
termByTerm(
    df=records, 
    termA='Keywords', 
    termB='Year', 
    sepA=';', 
    sepB=None, 
    ascendingA=None, ## unsorted
    ascendingB=None, 
    minmax=(2,3))
```

```python
##
## Keywords per Author
##
termByTerm(
    df=records, 
    termA='Authors', 
    termB='Keywords',
    sepA=',',
    sepB=';')
```

```python
##
## Keywords per year
##
termByTerm(
    df=records, 
    termA='Keywords', 
    termB='Year', 
    sepA=';', 
    sepB=None)
```

```python
##
## Document type per year
##
termByTerm(
    df=records, 
    termA='Document Type', 
    termB='Year', 
    sepA=None, 
    sepB=None)
```

```python
##
## keywords per Year
##
termByTerm(
    df=records, 
    termA='Keywords', 
    termB='Year', 
    sepA=';', 
    sepB=None, 
    ascendingA=True, 
    ascendingB=False)
```

```python
##
## Authors per Document Type
##
termByTerm(
    df=records, 
    termA='Authors', 
    termB='Document Type', 
    sepA=',', 
    sepB=None)
```
