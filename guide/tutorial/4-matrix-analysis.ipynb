# Matrix Analysis

In this part, a second level analysis of the bibliographical information is done.

```python
import pandas as pd
```

```python
records = pd.read_json('demo-keywords.json', orient='records', lines=True)
```

## Keywords per year

```python
from techMiner import termByTerm
```

```python
termByTerm(
    df=records, 
    termA='Keywords', 
    termB='Year', 
    sepA=';', 
    sepB=None).head()
```

```python
termByTerm( 
    df=records, 
    termA='Keywords', 
    termB='Year', 
    sepA=';', 
    sepB=None, 
    minmax=(2,3)).head()   ## minimum and maximum values in the matrix
```

```python
from techMiner import heatmap

heatmap(
    termByTerm( 
        df=records, 
        termA='Keywords', 
        termB='Year', 
        sepA=';', 
        sepB=None, 
        minmax=(2,8)),
    ascendingA=False, ascendingB=False)
```

```python
from techMiner import alt_heatmap

alt_heatmap(
    termByTerm( 
        df=records, 
        termA='Keywords', 
        termB='Year', 
        sepA=';', 
        sepB=None, 
        minmax=(2,8)))
```

```python
from techMiner import alt_circleplot

alt_circleplot(
    termByTerm( 
        df=records, 
        termA='Keywords', 
        termB='Year', 
        sepA=';', 
        sepB=None, 
        minmax=(2,8)))
```

## Autocorrelation

```python
from techMiner import correlation

correlation(
    df=records,
    termA='Keywords',
    sepA=';'
)
```

```python
alt_circleplot(
    correlation(
    df=records,
    termA='Keywords',
    sepA=';'))
```

```python
heatmap(
    correlation(
    df=records,
    termA='Keywords',
    sepA=';'))
```

## Crosscorrelation

```python
correlation(
    df=records,
    termA='Keywords',
    sepA=';',
    termB='Authors',
    sepB=','
)
```

```python
heatmap(
    correlation(
        df=records,
        termA='Keywords',
        sepA=';',
        termB='Authors',
        sepB=','))
```
