# Step 10: Cross-correlation

```python
import pandas as pd
from techminer import RecordsDataFrame

rdf = RecordsDataFrame(
    pd.read_json(
        'step-06.json', 
        orient='records', 
        lines=True))
```

In this part, a second level analysis of the bibliographical information is done.

## Autocorrelation

```python
rdf.autocorrelation(
    term='keywords (cleaned)',
    sep=';'
).to_matrix().head()
```

```python
rdf.autocorrelation(
    term='keywords (cleaned)',
    sep=';'
).circleplot_in_altair()
```

```python
rdf.autocorrelation(
    term='keywords (cleaned)',
    sep=';').heatmap()
```

## Crosscorrelation

```python
rdf.crosscorrelation(
    termA='keywords (cleaned)',
    sepA=';',
    termB='Authors',
    sepB=','
).head(10)
```

```python
rdf.crosscorrelation(
    termA='keywords (cleaned)',
    sepA=';',
    termB='Authors',
    sepB=','
).heatmap()
```

```python
rdf.crosscorrelation(
    termA='keywords (cleaned)',
    sepA=';',
    termB='Authors',
    sepB=','
).to_matrix()
```

```python
centers, clusters = rdf.crosscorrelation(
    termA='keywords (cleaned)',
    sepA=';',
    termB='Authors',
    sepB=','
).to_matrix().kmeans(n_clusters = 3)
```

```python
centers
```

```python
centers.transpose().heatmap()
```

## Factor analysis

```python
rdf.factor(
    term='keywords (cleaned)', 
    sep=';', 
    n_components=5, 
    N=40).head()
```

```python
rdf.factor(
    term='keywords (cleaned)', 
    sep=';', 
    n_components=5, 
    N=40).transpose().heatmap()
```
