# Step 10: Cross-correlation

```python
import pandas as pd
from techminer import RecordsDataFrame

rdf = RecordsDataFrame(
    pd.read_json(
        'step-06.json', 
        orient='records', 
        lines=True))
```

In this part, a second level analysis of the bibliographical information is done.

## Autocorrelation

```python
rdf.autocorrelation(
    term='keywords (cleaned)',
    sep=';'
).tomatrix().head()
```

```python
rdf.autocorrelation(
    term='keywords (cleaned)',
    sep=';'
).circlerel()
```

```python
rdf.autocorrelation(
    term='keywords (cleaned)',
    sep=';').heatmap()
```

```python
rdf.autocorrelation(
    term='keywords (cleaned)',
    sep=';').chord_diagram(minval=0.35)
```

```python
rdf.autocorrelation(
    term='Authors',
    sep=',').chord_diagram(minval=0.20)
```

## Crosscorrelation

```python
rdf.crosscorrelation(
    termA='keywords (cleaned)',
    sepA=';',
    termB='Authors',
    sepB=','
).head(10)
```

```python
rdf.crosscorrelation(
    termA='keywords (cleaned)',
    sepA=';',
    termB='Authors',
    sepB=','
).heatmap()
```

```python
rdf.crosscorrelation(
    termA='keywords (cleaned)',
    sepA=';',
    termB='Authors',
    sepB=','
).tomatrix()
```

```python
rdf.crosscorrelation(
    termA='keywords (cleaned)',
    sepA=';',
    termB='Authors',
    sepB=','
).sankey_plot(minval=0.2)
```

## Factor analysis

```python
rdf.factor_analysis(
    term='keywords (cleaned)', 
    sep=';', 
    n_components=5, 
    N=40).head(20)
```

```python
rdf.factor_analysis(
    term='keywords (cleaned)', 
    sep=';', 
    n_components=5, 
    N=40).heatmap()
```

```python
rdf.factor_analysis(
    term='keywords (cleaned)', 
    sep=';', 
    n_components=5, 
    N=40).sankey_plot(minval=0.3)
```
